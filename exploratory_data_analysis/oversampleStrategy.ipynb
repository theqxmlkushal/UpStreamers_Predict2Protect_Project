```python
# compare how many rows in df are exactly the same as rows in df1
common_cols = sorted(df.columns.intersection(df1.columns))

if not common_cols:
    print("No common columns between df and df1. Matching rows = 0")
else:
    # normalize NaNs and types so equal values compare properly
    df_sub = df[common_cols].fillna("__NA__").astype(str)
    df1_sub = df1[common_cols].fillna("__NA__").astype(str)

    a = df_sub.apply(tuple, axis=1)
    b = df1_sub.apply(tuple, axis=1)

    mask = a.isin(b)
    count = int(mask.sum())

    print("Common columns used for comparison:", common_cols)
    print("Matching rows in df that also appear in df1:", count)
    print(f"Total rows in df: {len(df)} (match rate: {count/len(df):.2%})")
```

```python
# compare company overlap between df and df_test to decide if LSTM training is feasible
train_set = set(df['company_name'].unique())
test_set = set(df_test['company_name'].unique())

common = sorted(train_set & test_set)
only_in_train = sorted(train_set - test_set)
only_in_test = sorted(test_set - train_set)

print(f"Companies in train: {len(train_set)}")
print(f"Companies in test:  {len(test_set)}")
print(f"Common companies:   {len(common)} "
    f"({len(common)/len(test_set):.2%} of test, {len(common)/len(train_set):.2%} of train)\n")

print("Sample common companies (first 20):", common[:20])
print("Sample only in train (first 10):", only_in_train[:10])
print("Sample only in test  (first 10):", only_in_test[:10])

# For sequence models you also care about number of years per company.
if common:
    train_years = df[df['company_name'].isin(common)].groupby('company_name')['fyear'].nunique()
    test_years = df_test[df_test['company_name'].isin(common)].groupby('company_name')['fyear'].nunique()
    combined = pd.DataFrame({'train_years': train_years, 'test_years': test_years}).fillna(0).astype(int)

    print("\nTime-series length stats for common companies (unique years):")
    print(combined.describe().transpose())

    # example filter: companies with enough history in train and at least one test year
    min_train_years = 5
    min_test_years = 1
    sufficient = combined[(combined['train_years'] >= min_train_years) & (combined['test_years'] >= min_test_years)]
    print(f"\nCommon companies with >= {min_train_years} train years and >= {min_test_years} test year: {len(sufficient)}")
    print(sufficient.head())
else:
    print("No common companies found.")
```

```python
# Visualize each column across df, df1, df_test
# Uses existing imports and dataframes (df, df1, df_test).
# For numeric-like columns: histogram + KDE. For categorical/low-cardinality: top-value bar chart.
# Plots one figure per column with up to 3 panels (df, df1, df_test).

datasets = [("df", df), ("df1", df1), ("df_test", df_test)]
all_cols = sorted(set(df.columns) | set(df1.columns) | set(df_test.columns))

for col in all_cols:
    fig, axes = plt.subplots(1, 3, figsize=(18, 4))
    for ax, (name, d) in zip(axes, datasets):
        if col not in d.columns:
            ax.axis("off")
            ax.set_title(f"{name}: missing")
            continue

        ser = d[col]
        # Try coerce to numeric to decide plot type
        ser_num = pd.to_numeric(ser, errors="coerce")
        n_non_na = ser_num.notna().sum()
        n_total = len(ser)

        # Heuristic: treat as numeric if a majority coerces to numeric and there are enough unique values
        if n_non_na >= max(10, 0.5 * n_total) and ser_num.nunique(dropna=True) > 8:
            sns.histplot(ser_num.dropna(), bins=30, kde=True, stat="density", ax=ax, color="C0")
            ax.set_xlabel(col)
            ax.set_ylabel("density")
            ax.set_title(f"{name} — numeric (n={n_non_na})")
        else:
            # categorical / low-cardinality
            vc = ser.fillna("__NA__").astype(str).value_counts().iloc[:20]
            sns.barplot(x=vc.values, y=vc.index, ax=ax, palette="muted")
            ax.set_xlabel("count")
            ax.set_title(f"{name} — top {len(vc)} categories")

    plt.suptitle(col, fontsize=14)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
from sklearn.metrics import f1_score, classification_report

# Compute F1 and macro F1 between sub (predictions) and df_ans (ground truth)


# align rows by company occurrence order to handle repeated company_name values
sub_aligned = sub.copy()
sub_aligned['__seq__'] = sub_aligned.groupby('company_name').cumcount()

df_ans_aligned = df_ans.copy()
df_ans_aligned['__seq__'] = df_ans_aligned.groupby('company_name').cumcount()

merged = pd.merge(
    sub_aligned.rename(columns={'status_label': 'pred'}),
    df_ans_aligned.rename(columns={'status_label': 'true'}),
    on=['company_name', '__seq__'],
    how='inner'
)

if len(merged) != len(sub) or len(merged) != len(df_ans):
    print(f"Warning: merged rows={len(merged)}, sub={len(sub)}, df_ans={len(df_ans)}")

y_true = merged['true']
y_pred = merged['pred']

# binary F1 (treat 'failed' as positive) and macro F1
f1_bin = f1_score(y_true, y_pred, average='binary', pos_label='failed')
f1_macro = f1_score(y_true, y_pred, average='macro')

print(f"F1 (binary, pos_label='failed'): {f1_bin:.6f}")
print(f"F1 (macro): {f1_macro:.6f}")

# optional brief report
print("\nPer-class F1:")
print(classification_report(y_true, y_pred, digits=4))
```
